```python
filename = glob.glob('./train/melanoma/*.jpg')
cnt = 0

for i in range(len(filename)):
    img = cv2.imread(filename[i])
    #img = cv2.circle(img, (350,220), 380, (0,0,0), 140)
    img = cv2.rectangle(img, (0,0), (720,480), (0,0,0), 120)
    cv2.imwrite('./train_final/train/melanoma/ISIC2_%d.jpg' % cnt, img)
    cnt += 1

print('Process Done!!')
```
